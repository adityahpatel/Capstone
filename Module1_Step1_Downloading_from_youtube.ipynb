## STEP3: Downloading 4 Professional matches

```python
# pro_1 video

link = 'https://www.youtube.com/watch?v=cHJ2jToKdLo' ## UPDATE

ydl_opts = {}

with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    info_dict = ydl.extract_info(link, download=False)
    video_title = info_dict.get('title', None)

path = f"/content/drive/MyDrive/CAPSTONE/CAPSTONE_FINAL/Module1_Video_downloading_and_Frame_extraction/video/'{video_title}.mp4" ## UPDATE

ydl_opts.update({'outtmpl':path})

with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    ydl.download([link])
```
