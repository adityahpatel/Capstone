## STEP3: Extracting all frames from 4 Professional matches to a specific folder

```python
# Loading Videos
# UPDATE following video paths

vid_pro_1="/content/drive/MyDrive/CAPSTONE/CAPSTONE_FINAL/Module1_Video_downloading_and_Frame_extraction/video/'Roland-Garros 2019 : Le résumé de la finale Dominic Thiem - Rafael Nadal.mp4"
vid_pro_3="/content/drive/MyDrive/CAPSTONE/CAPSTONE_FINAL/Module1_Video_downloading_and_Frame_extraction/video/'Djokovic Vs Nadal - Australian Open 2019 Final Highlights.mp4.mkv"
vid_pro_4="/content/drive/MyDrive/CAPSTONE/CAPSTONE_FINAL/Module1_Video_downloading_and_Frame_extraction/video/'US Open Extended Highlight: Novak Djokovic vs. Juan Martin Del Potro.mp4.mkv"
vid_pro_5="/content/drive/MyDrive/CAPSTONE/CAPSTONE_FINAL/Module1_Video_downloading_and_Frame_extraction/video/'Roger Federer v Marin Cilic highlights - Wimbledon 2017 final.mp4.mkv"
```

```python
vid_lst=[vid_pro_1,vid_pro_3,vid_pro_4,vid_pro_5]
vid_name_lst=['vid_pro_1','vid_pro_3','vid_pro_4','vid_pro_5']


for idx,val in enumerate(vid_lst):

  
  nam=vid_name_lst[idx]
  cap = cv2.VideoCapture(val)

  fps= int(cap.get(cv2.CAP_PROP_FPS))

  frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

  count=1
  while True:

      ret, frame = cap.read()

      name = '/content/drive/MyDrive/CAPSTONE/CAPSTONE_FINAL/Module1_Video_downloading_and_Frame_extraction/extracted_frames/professional/' +str(nam)+'_frame_'+ str(count) + '.jpg' ## UPDATE

      if ret:  
        # if count==1 or count%(round(fps/2,0))==0 or count==frame_count: # Save 1st, middle frames and last frame in video
        cv2.imwrite(name, frame)
        print ('Creating...' + name)
            
        count+=1

      else:
        break

      

  # Release all space and windows once done
  cap.release()
  cv2.destroyAllWindows()
```

```python
# Count number of frames extracted
folder_path='/content/drive/MyDrive/CAPSTONE/CAPSTONE_FINAL/Module1_Video_downloading_and_Frame_extraction/extracted_frames/professional' ## UPDATE

file_num=len(os.listdir(folder_path))
print(file_num)
```
